**IMPORTANDO BIBLIOTECA PANDAS**

```python
import pandas as pd #Importação padrão
```

**CARREGANDO DADOS** <br>
*É utilizado o "pd.read_csv" para ler e transformar os arquivos CSV em um DataFrame* <br>
*sep=";" Define o separador das colunas como ";"*<br>
*encoding="latin1" Define a codificação de caracteres usada no arquivo.( latin1 (ISO-8859-1) é usado para ler corretamente acentos e caracteres especiais como ç, ã, é. )* <br>
*header=1 Indica que o nome das colunas está na segunda linha do arquivo.*

```python
df2008 = pd.read_csv("despesa_ceaps_2008.csv", sep=";" , encoding="latin1", header=1)
df2009 = pd.read_csv("despesa_ceaps_2009.csv", sep=";" , encoding="latin1", header=1)
df2010 = pd.read_csv("despesa_ceaps_2010.csv", sep=";" , encoding="latin1", header=1)
df2011 = pd.read_csv("despesa_ceaps_2011.csv", sep=";" , encoding="latin1", header=1)
df2012 = pd.read_csv("despesa_ceaps_2012.csv", sep=";", encoding="latin1", header=1)
df2013 = pd.read_csv("despesa_ceaps_2013.csv", sep=";", encoding="latin1", header=1)
df2014 = pd.read_csv("despesa_ceaps_2014.csv", sep=";", encoding="latin1", header=1)
df2015 = pd.read_csv("despesa_ceaps_2015.csv", sep=";", encoding="latin1", header=1)
df2016 = pd.read_csv("despesa_ceaps_2016.csv", sep=";", encoding="latin1", header=1)
df2017 = pd.read_csv("despesa_ceaps_2017.csv", sep=";", encoding="latin1", header=1)
df2018 = pd.read_csv("despesa_ceaps_2018.csv", sep=";", encoding="latin1", header=1)
df2019 = pd.read_csv("despesa_ceaps_2019.csv", sep=";", encoding="latin1", header=1)
df2020 = pd.read_csv("despesa_ceaps_2020.csv", sep=";", encoding="latin1", header=1)
df2021 = pd.read_csv("despesa_ceaps_2021.csv", sep=";", encoding="latin1", header=1)
df2022 = pd.read_csv("despesa_ceaps_2022.csv", sep=";", encoding="latin1", header=1)
```

**JUNTANDO/EMPILHANDO TODOS OS DATASETS** <br>
*A função "pd.concat" faz esse serviço*

```python
df = pd.concat([df2008,df2009,df2010,df2011,df2012,df2013,df2014,df2014,df2016,df2017,df2018,df2019,df2020,df2021,df2022])
```

**RESETANDO O INDEX** <br>
*A maioria dos indices estavam duplicados, foi necessario reseta-los*

```python
df = df.reset_index(drop=True)
```

```python
df.index.duplicated().sum()
```

**TRANSFORMANDO A COLUNA "DATA" PARA O FORMATO CORRETO DE DATA**<br>
*Utilizando a função "pd.to_datetime" conseguimos fazer isso*<br>
*Detalhe que coloquei dayfirst como "True" para que iniciasse pelo dia, ja que o pandas reconhece as datas no padrão americano*

```python
df["DATA"] = pd.to_datetime(df["DATA"], dayfirst=True, errors="coerce")
```

```python
df.head()
```

**CONVERTENDO OS VALORES MONETARIOS**<br>
*Aqui, foi feito uma remoção de todos os "." da coluna;<br>
Troquei as vírgulas por ".";<br>
E converti para float os valores*

```python
df["VALOR_REEMBOLSADO"] = df["VALOR_REEMBOLSADO"].str.replace(".","")
df["VALOR_REEMBOLSADO"] = df["VALOR_REEMBOLSADO"].str.replace(",",".")
df["VALOR_REEMBOLSADO"] = df["VALOR_REEMBOLSADO"].astype(float)
```

```python
df["CNPJ_CPF"]
```

**Padronizar CNPJ_CPF**<br>
*Esse código remove tudo que não é número da coluna CNPJ_CPF*<br>
*O "r'\D'" é uma expressão regular (regex).<br>
\D = qualquer caractere que NÃO seja número.<br>
r'' = string raw (forma padrão de escrever regex no Python)<br>
Ou seja:<br>
\D → tudo que não é 0–9*

```python
df["CNPJ_CPF"] = df["CNPJ_CPF"].str.replace(r'\D', '', regex=True)
```

```python
df.head()
```

**Index para começar do 1**

```python
df.index = df.index + 1
```
